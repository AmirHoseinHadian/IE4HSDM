```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import pearsonr
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error as rmse
```

```python
recovery_2d_50 = pd.read_csv('../_Parameter_recovery/_Series/Series_2d_recovery_50.csv', index_col=0)
recovery_3d_50 = pd.read_csv('../_Parameter_recovery/_Series/Series_3d_recovery_50.csv', index_col=0)
recovery_4d_50 = pd.read_csv('../_Parameter_recovery/_Series/Series_4d_recovery_50.csv', index_col=0)

recovery_2d_100 = pd.read_csv('../_Parameter_recovery/_Series/Series_2d_recovery_100.csv', index_col=0)
recovery_3d_100 = pd.read_csv('../_Parameter_recovery/_Series/Series_3d_recovery_100.csv', index_col=0)
recovery_4d_100 = pd.read_csv('../_Parameter_recovery/_Series/Series_4d_recovery_100.csv', index_col=0)

recovery_2d = [recovery_2d_50, recovery_2d_100]
recovery_3d = [recovery_3d_50, recovery_3d_100]
recovery_4d = [recovery_4d_50, recovery_4d_100]
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.ylim([0, 2.2])
prms = 'threshold'
plt.plot([50, 100], [rmse(y_true=recovery_2d[i]['{}_true'.format(prms)],
                         y_pred=recovery_2d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '^');

plt.plot([50, 100], [rmse(y_true=recovery_3d[i]['{}_true'.format(prms)],
                         y_pred=recovery_3d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '^');

plt.plot([50], [rmse(y_true=recovery_4d[i]['{}_true'.format(prms)],
                         y_pred=recovery_4d[i]['{}_estimate'.format(prms)]) for i in range(1)],
         '^');

plt.subplot(132)
plt.ylim([0, 2.2])
prms = 'ndt'
plt.plot([50, 100], [rmse(y_true=recovery_2d[i]['{}_true'.format(prms)],
                         y_pred=recovery_2d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '^');

plt.plot([50, 100], [rmse(y_true=recovery_3d[i]['{}_true'.format(prms)],
                         y_pred=recovery_3d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '^');

plt.plot([50], [rmse(y_true=recovery_4d[i]['{}_true'.format(prms)],
                         y_pred=recovery_4d[i]['{}_estimate'.format(prms)]) for i in range(1)],
         '^');
```

```python
prms = 'threshold'

plt.plot([50, 100], [r2_score(y_true=recovery_2d[i]['{}_true'.format(prms)],
                         y_pred=recovery_2d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '^');

plt.plot([50, 100], [r2_score(y_true=recovery_3d[i]['{}_true'.format(prms)],
                         y_pred=recovery_3d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '^');

plt.plot([50], [r2_score(y_true=recovery_4d[i]['{}_true'.format(prms)],
                         y_pred=recovery_4d[i]['{}_estimate'.format(prms)]) for i in range(1)],
         '^');
```
