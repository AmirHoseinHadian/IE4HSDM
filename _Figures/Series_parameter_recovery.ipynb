```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import pearsonr
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error as rmse
```

```python
recovery_2d_50 = pd.read_csv('../_Parameter_recovery/_Series/Series_2d_recovery_50.csv', index_col=0)
recovery_3d_50 = pd.read_csv('../_Parameter_recovery/_Series/Series_3d_recovery_50.csv', index_col=0)
recovery_4d_50 = pd.read_csv('../_Parameter_recovery/_Series/Series_4d_recovery_50.csv', index_col=0)

recovery_2d_100 = pd.read_csv('../_Parameter_recovery/_Series/Series_2d_recovery_100.csv', index_col=0)
recovery_3d_100 = pd.read_csv('../_Parameter_recovery/_Series/Series_3d_recovery_100.csv', index_col=0)
recovery_4d_100 = pd.read_csv('../_Parameter_recovery/_Series/Series_4d_recovery_100.csv', index_col=0)

recovery_2d_250 = pd.read_csv('../_Parameter_recovery/_Series/Series_2d_recovery_250.csv', index_col=0)
# recovery_3d_250 = pd.read_csv('../_Parameter_recovery/_Series/Series_3d_recovery_250.csv', index_col=0)
# recovery_4d_250 = pd.read_csv('../_Parameter_recovery/_Series/Series_4d_recovery_250.csv', index_col=0)

recovery_2d = [recovery_2d_50, recovery_2d_100, recovery_2d_250]
recovery_3d = [recovery_3d_50, recovery_3d_100]
recovery_4d = [recovery_4d_50, recovery_4d_100]
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.ylim([0, 1])
prms = 'threshold'
plt.plot([50, 100, 250], 
         [rmse(y_true=recovery_2d[i]['{}_true'.format(prms)],
               y_pred=recovery_2d[i]['{}_estimate'.format(prms)]) for i in range(3)],
         '--^');

plt.plot([50, 100], [rmse(y_true=recovery_3d[i]['{}_true'.format(prms)],
                         y_pred=recovery_3d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '--^');

plt.plot([50, 100], [rmse(y_true=recovery_4d[i]['{}_true'.format(prms)],
                         y_pred=recovery_4d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '--^');

plt.subplot(132)
plt.ylim([0, 1])
prms = 'ndt'
plt.plot([50, 100, 250], [rmse(y_true=recovery_2d[i]['{}_true'.format(prms)],
                         y_pred=recovery_2d[i]['{}_estimate'.format(prms)]) for i in range(3)],
         '--^');

plt.plot([50, 100], [rmse(y_true=recovery_3d[i]['{}_true'.format(prms)],
                         y_pred=recovery_3d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '--^');

plt.plot([50, 100], [rmse(y_true=recovery_4d[i]['{}_true'.format(prms)],
                         y_pred=recovery_4d[i]['{}_estimate'.format(prms)]) for i in range(2)],
         '--^');

plt.subplot(133)
plt.ylim([0, 1.2])
mu_true = [np.append([recovery_2d[i]['mu1_true'], 
                      recovery_2d[i]['mu2_true']], 0) for i in range(3)]
mu_estimate = [np.append([recovery_2d[i]['mu1_estimate'], 
                          recovery_2d[i]['mu2_estimate']], 0) for i in range(3)]
plt.plot([50, 100, 250], [rmse(mu_true[i],mu_estimate[i]) for i in range(3)],
         '--^', label='2D');

mu_true = [np.append([recovery_3d[i]['mu1_true'], 
                      recovery_3d[i]['mu2_true'], 
                      recovery_3d[i]['mu3_true']], 0) for i in range(2)]
mu_estimate = [np.append([recovery_3d[i]['mu1_estimate'], 
                          recovery_3d[i]['mu2_estimate'], 
                          recovery_3d[i]['mu3_estimate']], 0) for i in range(2)]
plt.plot([50, 100], [rmse(mu_true[i], mu_estimate[i]) for i in range(2)],
         '--^', label='3D');

mu_true = [np.append([recovery_4d[i]['mu1_true'], 
                      recovery_4d[i]['mu2_true'], 
                      recovery_4d[i]['mu3_true'],
                      recovery_4d[i]['mu4_true']], 0) for i in range(2)]
mu_estimate = [np.append([recovery_4d[i]['mu1_estimate'], 
                          recovery_4d[i]['mu2_estimate'], 
                          recovery_4d[i]['mu3_estimate'],
                          recovery_4d[i]['mu4_estimate']], 0) for i in range(2)]
plt.plot([50, 100], [rmse(mu_true[i], mu_estimate[i]) for i in range(2)],
         '--^', label='4D');

plt.legend();
```
