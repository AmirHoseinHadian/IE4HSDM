```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error as rmse
```

```python
recovery_2d = pd.read_csv('_Recovery_data/IE_2d_recovery_linear_500_02.csv', index_col=0)
recovery_3d = pd.read_csv('_Recovery_data/IE_3d_recovery_linear_500_02.csv', index_col=0)
recovery_4d = pd.read_csv('_Recovery_data/IE_4d_recovery_linear_500_02.csv', index_col=0)

mu_2d_true = np.append([recovery_2d['mu1_true'], 
                        recovery_2d['mu2_true']], 0)
mu_2d_estimate = np.append([recovery_2d['mu1_estimate'], 
                            recovery_2d['mu2_estimate']], 0)

mu_3d_true = np.append([recovery_3d['mu1_true'],
                        recovery_3d['mu2_true'],
                        recovery_3d['mu3_true']], 0)
mu_3d_estimate = np.append([recovery_3d['mu1_estimate'],
                            recovery_3d['mu2_estimate'],
                            recovery_3d['mu3_estimate']], 0)

mu_4d_true = np.append([recovery_4d['mu1_true'],
                        recovery_4d['mu2_true'],
                        recovery_4d['mu3_true'],
                        recovery_4d['mu4_true']], 0)
mu_4d_estimate = np.append([recovery_4d['mu1_estimate'],
                            recovery_4d['mu2_estimate'],
                            recovery_4d['mu3_estimate'],
                            recovery_4d['mu4_estimate']], 0)
```

```python
print(rmse(y_true=recovery_2d['b0_true'], y_pred=recovery_2d['b0_estimate']))
print(rmse(y_true=recovery_2d['lambda_true'], y_pred=recovery_2d['lambda_estimate']))
print(rmse(y_true=recovery_2d['ndt_true'], y_pred=recovery_2d['ndt_estimate']))
print(rmse(y_true=mu_2d_true, y_pred=mu_2d_estimate))

print('============')

print(rmse(y_true=recovery_3d['b0_true'], y_pred=recovery_3d['b0_estimate']))
print(rmse(y_true=recovery_3d['lambda_true'], y_pred=recovery_3d['lambda_estimate']))
print(rmse(y_true=recovery_3d['ndt_true'], y_pred=recovery_3d['ndt_estimate']))
print(rmse(y_true=mu_3d_true, y_pred=mu_3d_estimate))

print('============')

print(rmse(y_true=recovery_4d['b0_true'], y_pred=recovery_4d['b0_estimate']))
print(rmse(y_true=recovery_4d['lambda_true'], y_pred=recovery_4d['lambda_estimate']))
print(rmse(y_true=recovery_4d['ndt_true'], y_pred=recovery_4d['ndt_estimate']))
print(rmse(y_true=mu_4d_true, y_pred=mu_4d_estimate))
```

```python
print(r2_score(y_true=recovery_2d['b0_true'], y_pred=recovery_2d['b0_estimate']))
print(r2_score(y_true=recovery_2d['lambda_true'], y_pred=recovery_2d['lambda_estimate']))
print(r2_score(y_true=recovery_2d['ndt_true'], y_pred=recovery_2d['ndt_estimate']))
print(r2_score(y_true=mu_2d_true, y_pred=mu_2d_estimate))

print('============')

print(r2_score(y_true=recovery_3d['b0_true'], y_pred=recovery_3d['b0_estimate']))
print(r2_score(y_true=recovery_3d['lambda_true'], y_pred=recovery_3d['lambda_estimate']))
print(r2_score(y_true=recovery_3d['ndt_true'], y_pred=recovery_3d['ndt_estimate']))
print(r2_score(y_true=mu_3d_true, y_pred=mu_3d_estimate))

print('============')

print(r2_score(y_true=recovery_4d['b0_true'], y_pred=recovery_4d['b0_estimate']))
print(r2_score(y_true=recovery_4d['lambda_true'], y_pred=recovery_4d['lambda_estimate']))
print(r2_score(y_true=recovery_4d['ndt_true'], y_pred=recovery_4d['ndt_estimate']))
print(r2_score(y_true=mu_4d_true, y_pred=mu_4d_estimate))
```

```python
print(pearsonr(recovery_2d['b0_true'], recovery_2d['b0_estimate'])[0])
print(pearsonr(recovery_2d['lambda_true'], recovery_2d['lambda_estimate'])[0])
print(pearsonr(recovery_2d['ndt_true'], recovery_2d['ndt_estimate'])[0])
print(pearsonr(mu_2d_true, mu_2d_estimate)[0])

print('============')

print(pearsonr(recovery_3d['b0_true'], recovery_3d['b0_estimate'])[0])
print(pearsonr(recovery_3d['lambda_true'], recovery_3d['lambda_estimate'])[0])
print(pearsonr(recovery_3d['ndt_true'], recovery_3d['ndt_estimate'])[0])
print(pearsonr(mu_3d_true, mu_3d_estimate)[0])

print('============')

print(pearsonr(recovery_4d['b0_true'], recovery_4d['b0_estimate'])[0])
print(pearsonr(recovery_4d['lambda_true'], recovery_4d['lambda_estimate'])[0])
print(pearsonr(recovery_4d['ndt_true'], recovery_4d['ndt_estimate'])[0])
print(pearsonr(mu_4d_true, mu_4d_estimate)[0])
```

```python
plt.rc('axes', titlesize=15)     # fontsize of the axes title
plt.rc('axes', labelsize=18)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
plt.rc('legend', fontsize=15)    # legend fontsize

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(25, 18))

axs[0, 0].plot(np.linspace(2, 5, 100), np.linspace(2, 5, 100), c='red')
axs[0, 0].scatter(x=recovery_2d['b0_true'], y=recovery_2d['b0_estimate'], c='green', alpha=0.3);
axs[0, 0].set_xlabel(r'True starting threshold ($b_0$)')
axs[0, 0].set_ylabel('2D\n\nEstimated starting threshold ($b_0$)')
R2 = r2_score(y_true=recovery_2d['b0_true'], y_pred=recovery_2d['b0_estimate'])
roh = pearsonr(recovery_2d['b0_true'], recovery_2d['b0_estimate'])[0]
axs[0, 0].text(2, 4.8, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[0, 0].text(4, 2, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[0, 1].plot(np.linspace(0, 2.2, 100), np.linspace(0, 2.2, 100), c='red')
axs[0, 1].scatter(x=recovery_2d['lambda_true'], y=recovery_2d['lambda_estimate'], c='green', alpha=0.3);
axs[0, 1].set_xlabel(r'True slope ($\lambda$)')
axs[0, 1].set_ylabel('Estimated slope ($\lambda$)')
R2 = r2_score(y_true=recovery_2d['lambda_true'], y_pred=recovery_2d['lambda_estimate'])
roh = pearsonr(recovery_2d['lambda_true'], recovery_2d['lambda_estimate'])[0]
axs[0, 1].text(0, 2.1, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[0, 1].text(1.5, 0, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[0, 2].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[0, 2].scatter(x=recovery_2d['ndt_true'], y=recovery_2d['ndt_estimate'], c='green', alpha=0.3);
axs[0, 2].set_xlabel(r'True non-decision time ($t_0$)')
axs[0, 2].set_ylabel(r'Estimated non-decision time ($t_0$)')
R2 = r2_score(y_true=recovery_2d['ndt_true'], y_pred=recovery_2d['ndt_estimate'])
roh = pearsonr(recovery_2d['ndt_true'], recovery_2d['ndt_estimate'])[0]
axs[0, 2].text(0.1, 0.93, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[0, 2].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[0, 3].plot(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100), c='red')
axs[0, 3].scatter(x=mu_2d_true, y=mu_2d_estimate, c='green', alpha=0.3);
axs[0, 3].set_xlabel(r'True drift rate ($\mu$)')
axs[0, 3].set_ylabel(r'Estimated drift rate ($\mu$)')
R2 = r2_score(y_true=mu_2d_true, y_pred=mu_2d_estimate)
roh = pearsonr(mu_2d_true, mu_2d_estimate)[0]
axs[0, 3].text(-3, 2.8, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[0, 3].text(1.5, -3, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

############

axs[1, 0].plot(np.linspace(2, 5, 100), np.linspace(2, 5, 100), c='red')
axs[1, 0].scatter(x=recovery_3d['b0_true'], y=recovery_3d['b0_estimate'], c='green', alpha=0.3);
axs[1, 0].set_xlabel(r'True starting threshold ($b_0$)')
axs[1, 0].set_ylabel('3D\n\nEstimated starting threshold ($b_0$)')
R2 = r2_score(y_true=recovery_3d['b0_true'], y_pred=recovery_3d['b0_estimate'])
roh = pearsonr(recovery_3d['b0_true'], recovery_3d['b0_estimate'])[0]
axs[1, 0].text(2, 4.8, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[1, 0].text(4, 2, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[1, 1].plot(np.linspace(0, 2.2, 100), np.linspace(0, 2.2, 100), c='red')
axs[1, 1].scatter(x=recovery_3d['lambda_true'], y=recovery_3d['lambda_estimate'], c='green', alpha=0.3);
axs[1, 1].set_xlabel(r'True slope ($\lambda$)')
axs[1, 1].set_ylabel('Estimated slope ($\lambda$)')
R2 = r2_score(y_true=recovery_3d['lambda_true'], y_pred=recovery_3d['lambda_estimate'])
roh = pearsonr(recovery_3d['lambda_true'], recovery_3d['lambda_estimate'])[0]
axs[1, 1].text(0, 2.1, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[1, 1].text(1.5, 0, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[1, 2].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[1, 2].scatter(x=recovery_3d['ndt_true'], y=recovery_3d['ndt_estimate'], c='green', alpha=0.3);
axs[1, 2].set_xlabel(r'True non-decision time ($t_0$)')
axs[1, 2].set_ylabel(r'Estimated non-decision time ($t_0$)')
R2 = r2_score(y_true=recovery_3d['ndt_true'], y_pred=recovery_3d['ndt_estimate'])
roh = pearsonr(recovery_3d['ndt_true'], recovery_3d['ndt_estimate'])[0]
axs[1, 2].text(0.1, 0.93, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[1, 2].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[1, 3].plot(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100), c='red')
axs[1, 3].scatter(x=mu_3d_true, y=mu_3d_estimate, c='green', alpha=0.3);
axs[1, 3].set_xlabel(r'True drift rate ($\mu$)')
axs[1, 3].set_ylabel(r'Estimated drift rate ($\mu$)')
R2 = r2_score(y_true=mu_3d_true, y_pred=mu_3d_estimate)
roh = pearsonr(mu_3d_true, mu_3d_estimate)[0]
axs[1, 3].text(-3, 2.8, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[1, 3].text(1.5, -3, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)


############

axs[2, 0].plot(np.linspace(2, 5, 100), np.linspace(2, 5, 100), c='red')
axs[2, 0].scatter(x=recovery_4d['b0_true'], y=recovery_4d['b0_estimate'], c='green', alpha=0.3);
axs[2, 0].set_xlabel(r'True starting threshold ($b_0$)')
axs[2, 0].set_ylabel('4D\n\nEstimated starting threshold ($b_0$)')
R2 = r2_score(y_true=recovery_4d['b0_true'], y_pred=recovery_4d['b0_estimate'])
roh = pearsonr(recovery_4d['b0_true'], recovery_4d['b0_estimate'])[0]
axs[2, 0].text(2, 4.8, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[2, 0].text(4, 2, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[2, 1].plot(np.linspace(0, 2.2, 100), np.linspace(0, 2.2, 100), c='red')
axs[2, 1].scatter(x=recovery_4d['lambda_true'], y=recovery_4d['lambda_estimate'], c='green', alpha=0.3);
axs[2, 1].set_xlabel(r'True slope ($\lambda$)')
axs[2, 1].set_ylabel('Estimated slope ($\lambda$)')
R2 = r2_score(y_true=recovery_4d['lambda_true'], y_pred=recovery_4d['lambda_estimate'])
roh = pearsonr(recovery_4d['lambda_true'], recovery_4d['lambda_estimate'])[0]
axs[2, 1].text(0, 2.1, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[2, 1].text(1.5, 0, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[2, 2].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[2, 2].scatter(x=recovery_4d['ndt_true'], y=recovery_4d['ndt_estimate'], c='green', alpha=0.3);
axs[2, 2].set_xlabel(r'True non-decision time ($t_0$)')
axs[2, 2].set_ylabel(r'Estimated non-decision time ($t_0$)')
R2 = r2_score(y_true=recovery_4d['ndt_true'], y_pred=recovery_4d['ndt_estimate'])
roh = pearsonr(recovery_4d['ndt_true'], recovery_4d['ndt_estimate'])[0]
axs[2, 2].text(0.1, 0.93, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[2, 2].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18)

axs[2, 3].plot(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100), c='red')
axs[2, 3].scatter(x=mu_4d_true, y=mu_4d_estimate, c='green', alpha=0.3);
axs[2, 3].set_xlabel(r'True drift rate ($\mu$)')
axs[2, 3].set_ylabel(r'Estimated drift rate ($\mu$)');
R2 = r2_score(y_true=mu_4d_true, y_pred=mu_4d_estimate)
roh = pearsonr(mu_4d_true, mu_4d_estimate)[0]
axs[2, 3].text(-3, 2.8, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 18)
axs[2, 3].text(1.5, -3, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 18);

plt.savefig('../_01_Figures/ie_recovery_linear.pdf', 
            transparent=True, 
            dpi=600, 
            bbox_inches='tight');
```
