```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from scipy.stats import pearsonr
from sklearn.metrics import r2_score
```

```python
recovery_2d = pd.read_csv('_Recovery_data/IE_2d_recovery_dvar.csv', index_col=0)
recovery_3d = pd.read_csv('_Recovery_data/IE_3d_recovery_dvar.csv', index_col=0)
recovery_4d = pd.read_csv('_Recovery_data/IE_4d_recovery_dvar.csv', index_col=0)

mu_2d_true = np.append([recovery_2d['mu1_true'], 
                        recovery_2d['mu2_true']], 0)
mu_2d_estimate = np.append([recovery_2d['mu1_estimate'], 
                            recovery_2d['mu2_estimate']], 0)

mu_3d_true = np.append([recovery_3d['mu1_true'],
                        recovery_3d['mu2_true'],
                        recovery_3d['mu3_true']], 0)
mu_3d_estimate = np.append([recovery_3d['mu1_estimate'],
                            recovery_3d['mu2_estimate'],
                            recovery_3d['mu3_estimate']], 0)

mu_4d_true = np.append([recovery_4d['mu1_true'],
                        recovery_4d['mu2_true'],
                        recovery_4d['mu3_true'],
                        recovery_4d['mu4_true']], 0)
mu_4d_estimate = np.append([recovery_4d['mu1_estimate'],
                            recovery_4d['mu2_estimate'],
                            recovery_4d['mu3_estimate'],
                            recovery_4d['mu4_estimate']], 0)
```

```python
plt.rc('axes', titlesize=15)     # fontsize of the axes title
plt.rc('axes', labelsize=18)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=15)    # fontsize of the tick labels
plt.rc('ytick', labelsize=15)    # fontsize of the tick labels
plt.rc('legend', fontsize=15)    # legend fontsize

fig, axs = plt.subplots(nrows=3, ncols=4, figsize=(24, 17))

axs[0, 0].plot(np.linspace(0.5, 3.5, 100), np.linspace(0.5, 3.5, 100), c='red')
axs[0, 0].scatter(x=recovery_2d['threshold_true'], y=recovery_2d['threshold_estimate'], c='green', alpha=0.3);
axs[0, 0].set_xlabel(r'True threshold ($b$)')
axs[0, 0].set_ylabel('2D\n\nEstimated threshold ($b$)')
R2 = r2_score(y_true=recovery_2d['threshold_true'], y_pred=recovery_2d['threshold_estimate'])
roh = pearsonr(recovery_2d['threshold_true'], recovery_2d['threshold_estimate'])[0]
axs[0, 0].text(0.5, 3.25, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[0, 0].text(2.5, 0.5, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[0, 1].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[0, 1].scatter(x=recovery_2d['ndt_true'], y=recovery_2d['ndt_estimate'], c='green', alpha=0.3);
axs[0, 1].set_xlabel(r'True non-decision time ($t_0$)')
axs[0, 1].set_ylabel(r'Estimated non-decision time ($t_0$)')
R2 = r2_score(y_true=recovery_2d['ndt_true'], y_pred=recovery_2d['ndt_estimate'])
roh = pearsonr(recovery_2d['ndt_true'], recovery_2d['ndt_estimate'])[0]
axs[0, 1].text(0.1, 0.93, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[0, 1].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[0, 2].plot(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100), c='red')
axs[0, 2].scatter(x=mu_2d_true, y=mu_2d_estimate, c='green', alpha=0.3);
axs[0, 2].set_xlabel(r'True drift rate ($\mu_i$)')
axs[0, 2].set_ylabel(r'Estimated drift rate ($\mu_i$)')
R2 = r2_score(y_true=mu_2d_true, y_pred=mu_2d_estimate)
roh = pearsonr(mu_2d_true, mu_2d_estimate)[0]
axs[0, 2].text(-3, 2.5, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[0, 2].text(1, -3, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[0, 3].scatter(recovery_2d['eta_true'], recovery_2d['eta_estimate'], c='green', alpha=0.3)
axs[0, 3].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[0, 3].set_xlabel(r'True drift variability ($\eta$)')
axs[0, 3].set_ylabel(r'Estimated drift variability ($\eta$)');
R2 = r2_score(y_true=recovery_2d['eta_true'], y_pred=recovery_2d['eta_estimate'])
roh = pearsonr(recovery_2d['eta_true'], recovery_2d['eta_estimate'])[0]
axs[0, 3].text(0.1, 0.95, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[0, 3].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20);


############

axs[1, 0].plot(np.linspace(0.5, 3.5, 100), np.linspace(0.5, 3.5, 100), c='red')
axs[1, 0].scatter(x=recovery_3d['threshold_true'], y=recovery_3d['threshold_estimate'], c='green', alpha=0.3);
axs[1, 0].set_xlabel(r'True threshold ($b$)')
axs[1, 0].set_ylabel('3D\n\nEstimated threshold ($b$)')
R2 = r2_score(y_true=recovery_3d['threshold_true'], y_pred=recovery_3d['threshold_estimate'])
roh = pearsonr(recovery_3d['threshold_true'], recovery_3d['threshold_estimate'])[0]
axs[1, 0].text(0.5, 3.25, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[1, 0].text(2.5, 0.5, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[1, 1].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[1, 1].scatter(x=recovery_3d['ndt_true'], y=recovery_3d['ndt_estimate'], c='green', alpha=0.3);
axs[1, 1].set_xlabel(r'True non-decision time ($t_0$)')
axs[1, 1].set_ylabel('Estimated non-decision time ($t_0$)')
R2 = r2_score(y_true=recovery_3d['ndt_true'], y_pred=recovery_3d['ndt_estimate'])
roh = pearsonr(recovery_3d['ndt_true'], recovery_3d['ndt_estimate'])[0]
axs[1, 1].text(0.1, 0.93, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[1, 1].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[1, 2].plot(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100), c='red')
axs[1, 2].scatter(x=mu_3d_true, y=mu_3d_estimate, c='green', alpha=0.3);
axs[1, 2].set_xlabel(r'True drift rate ($\mu_i$)')
axs[1, 2].set_ylabel(r'Estimated drift rate ($\mu_i$)')
R2 = r2_score(y_true=mu_3d_true, y_pred=mu_3d_estimate)
roh = pearsonr(mu_3d_true, mu_3d_estimate)[0]
axs[1, 2].text(-3, 2.5, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[1, 2].text(1, -3, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[1, 3].scatter(recovery_3d['eta_true'], recovery_3d['eta_estimate'], c='green', alpha=0.3)
axs[1, 3].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red');
axs[1, 3].set_xlabel(r'True drift variability ($\eta$)');
axs[1, 3].set_ylabel(r'Estimated drift variability ($\eta$)');
R2 = r2_score(y_true=recovery_3d['eta_true'], y_pred=recovery_3d['eta_estimate'])
roh = pearsonr(recovery_3d['eta_true'], recovery_3d['eta_estimate'])[0]
axs[1, 3].text(0.1, 0.95, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[1, 3].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20);


############

axs[2, 0].plot(np.linspace(0.5, 3.5, 100), np.linspace(0.5, 3.5, 100), c='red')
axs[2, 0].scatter(x=recovery_4d['threshold_true'], y=recovery_4d['threshold_estimate'], c='green', alpha=0.3);
axs[2, 0].set_xlabel(r'True threshold ($b$)')
axs[2, 0].set_ylabel('4D\n\nEstimated threshold ($b$)')
R2 = r2_score(y_true=recovery_4d['threshold_true'], y_pred=recovery_4d['threshold_estimate'])
roh = pearsonr(recovery_4d['threshold_true'], recovery_4d['threshold_estimate'])[0]
axs[2, 0].text(0.5, 3.25, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[2, 0].text(2.5, 0.5, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[2, 1].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[2, 1].scatter(x=recovery_4d['ndt_true'], y=recovery_4d['ndt_estimate'], c='green', alpha=0.3);
axs[2, 1].set_xlabel(r'True non-decision time ($t_0$)')
axs[2, 1].set_ylabel(r'Estimated non-decision time ($t_0$)')
R2 = r2_score(y_true=recovery_4d['ndt_true'], y_pred=recovery_4d['ndt_estimate'])
roh = pearsonr(recovery_4d['ndt_true'], recovery_4d['ndt_estimate'])[0]
axs[2, 1].text(0.1, 0.93, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[2, 1].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20)

axs[2, 2].plot(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100), c='red')
axs[2, 2].scatter(x=mu_4d_true, y=mu_4d_estimate, c='green', alpha=0.3);
axs[2, 2].set_xlabel(r'True drift rate ($\mu_i$)')
axs[2, 2].set_ylabel(r'Estimated drift rate ($\mu_i$)');
R2 = r2_score(y_true=mu_4d_true, y_pred=mu_4d_estimate)
roh = pearsonr(mu_4d_true, mu_4d_estimate)[0]
axs[2, 2].text(-3, 2.5, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[2, 2].text(1, -3, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20);

axs[2, 3].scatter(recovery_4d['eta_true'], recovery_4d['eta_estimate'], c='green', alpha=0.3)
axs[2, 3].plot(np.linspace(0.1, 1, 100), np.linspace(0.1, 1, 100), c='red')
axs[2, 3].set_xlabel(r'True drift variability ($\eta$)')
axs[2, 3].set_ylabel(r'Estimated drift variability ($\eta$)');
R2 = r2_score(y_true=recovery_4d['eta_true'], y_pred=recovery_4d['eta_estimate'])
roh = pearsonr(recovery_4d['eta_true'], recovery_4d['eta_estimate'])[0]
axs[2, 3].text(0.1, 0.95, r'$R^2 = {}$'.format(str(R2)[:4]), fontsize = 20)
axs[2, 3].text(0.7, 0.1, r'$\rho = {}$'.format(str(roh)[:4]), fontsize = 20);

plt.savefig('../../../_01_Figures/dvar_recovery.pdf', 
            transparent=True, 
            dpi=600, 
            bbox_inches='tight');
```
